```python
# ============================================================
# 승복 파생컬럼 (퍼널 전환 분석용)
# 기준: received 기준 1행 (customer_id + offer_id + time_received)
# ============================================================

# has_viewed: received → viewed 여부 (1=열람, 0=미열람)
funnel['has_viewed'] = funnel['time_viewed'].notna().astype(int)

# has_completed_rc: received → completed 여부 (1=전환, 0=미전환)
funnel['has_completed_rc'] = funnel['time_completed_rc'].notna().astype(int)

# has_completed_after_view: viewed → completed 여부 (1=열람 후 전환, 0=미전환)
funnel['has_completed_after_view'] = funnel['time_completed_vc'].notna().astype(int)

# status: 퍼널 단계별 상태 구분 (전환 / 미열람 / 열람후미전환)
funnel['status'] = np.select(
    [
        funnel['has_completed_rc'] == 1,
        funnel['has_viewed'] == 0,
        (funnel['has_viewed'] == 1) & (funnel['has_completed_rc'] == 0)
    ],
    [
        'converted',
        'not_viewed',
        'viewed_not_converted'
    ],
    default='other'
)

# merge용 컬럼
df_sb_features = funnel[
    [
        'customer_id',
        'offer_id',
        'time_received',
        'has_viewed',
        'has_completed_rc',
        'has_completed_after_view',
        'status'
    ]
].copy()
```
