```python
# 1. 컬럼명 통일(merge할때 쉽게)
profile = profile.rename(columns={"id": "customer_id"})
transcript = transcript.rename(columns={"person": "customer_id"})
portfolio = portfolio.rename(columns={"id": "offer_id"})

# 2. value 컬럼이 문자열형태로 되어있어 딕셔너리로 변환
transcript["value"] = transcript["value"].apply(
    lambda x: ast.literal_eval(x) if isinstance(x, str) else x
)

# 3. transcript에서 (value에서 필요컬럼 3개 추출)
transcript["offer_id"] = transcript["value"].apply(
    lambda x: x.get("offer id") or x.get("offer_id") if isinstance(x, dict) else None
)
transcript["amount"] = transcript["value"].apply(
    lambda x: x.get("amount") if isinstance(x, dict) else None
)
transcript["reward_value"] = transcript["value"].apply(
    lambda x: x.get("reward") if isinstance(x, dict) else None
)

# 4. merge로 통합
df_merge = transcript.merge(profile, on="customer_id", how="left")
df_merge = df_merge.merge(portfolio, on="offer_id", how="left")

# 5. 확인
print("shape:", df_merge.shape)
print("결측치 확인:", df_merge.isna().sum())
```
